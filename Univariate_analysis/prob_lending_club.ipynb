```python
"""program to explore loan data from the lending club"""

import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats

#read loan data and load to pandas dataframe, removing rows with incomplete data
loansData = pd.read_csv('https://spark-public.s3.amazonaws.com/dataanalysis/loansData.csv')
loansData.dropna(inplace=True)
```

```python
#print first five rows to take a look at data
print(loansData.head())
```

```python
#Describe the data using describe(). This is not exactly the same as the str function in R, so bit less informative
print(loansData.describe())
```

```python
%pylab inline
```

```python
#create and save a box plot showing  amount requested and amount funded by investors
```

```python
plt.figure()
loansData.boxplot(column=["Amount.Requested", "Amount.Funded.By.Investors"], return_type='dict')
plt.ylabel('Amount in dollars')
plt.savefig("boxplot1.png")
plt.show()
    
```

```python
#create and save histogram of amount requested and amount funded by investors
```

```python
plt.figure(figsize=(14, 4))
plt.subplot(1, 2, 1)
plt.hist(loansData["Amount.Requested"].values)
plt.title('Amount Requested')
plt.ylabel('Amount in dollars')

plt.subplot(1, 2, 2)
plt.hist(loansData["Amount.Funded.By.Investors"].values)
plt.title('Amount funded by investors')
plt.savefig("histogram1.png")
plt.show()
```

```python
#create and save QQ-plot of amount requested and amount funded by investors
```

```python
fig = plt.figure(figsize=(12, 4))
g1 = fig.add_subplot(1, 2, 1)
graph = stats.probplot(loansData["Amount.Requested"], dist="norm", plot=g1)
plt.title("QQ-plot:Amount Requested")  
plt.ylabel("Values in dollars")

g2 = fig.add_subplot(1, 2, 2)
graph = stats.probplot(loansData["Amount.Funded.By.Investors"], dist="norm", plot=g2)
plt.title("QQ-plot:Amount Funded By Investors")
plt.ylabel(" ")
plt.savefig("qqplot1.png")
plt.show()
```
